```python
import pandas as pd
import re

# =====================================================
# CONFIGURATION
# =====================================================

INPUT_CSV = "user_content.csv"
OUTPUT_CSV = "output_with_wordcount.csv"

# =====================================================
# WORD COUNT FUNCTION
# =====================================================

# URLs
URL_PATTERN = re.compile(
    r"(?:https?://[^\s]+|ftp://[^\s]+|www\.[^\s]+)",
    flags=re.IGNORECASE
)

# Emails
EMAIL_PATTERN = re.compile(
    r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b",
    flags=re.IGNORECASE
)

# Normal words/numbers with internal hyphens
WORD_PATTERN = re.compile(
    r"\b[A-Za-z0-9]+(?:-[A-Za-z0-9]+)*\b"
)

def count_words(text):

    if pd.isna(text):
        return 0

    text = str(text)

    # Count URLs
    urls = URL_PATTERN.findall(text)
    url_count = len(urls)

    text = URL_PATTERN.sub(" ", text)

    # Count emails
    emails = EMAIL_PATTERN.findall(text)
    email_count = len(emails)

    text = EMAIL_PATTERN.sub(" ", text)

    # Count remaining words
    normal_words = WORD_PATTERN.findall(text)

    return url_count + email_count + len(normal_words)


# =====================================================
# LOAD DATA
# =====================================================

df = pd.read_csv(INPUT_CSV)

if "original_post" not in df.columns:
    raise ValueError("Column 'original_post' not found.")

# =====================================================
# COMPUTE WORD COUNTS
# =====================================================

df["word_count"] = df["original_post"].apply(count_words)

# =====================================================
# SAVE RESULTS
# =====================================================

df.to_csv(OUTPUT_CSV, index=False)

print(f"Processed {len(df)} rows.")
print(f"Output saved to: {OUTPUT_CSV}")

# Show sample
print(df[["original_post", "word_count"]].head())
```

```python
# Word count for posts > 500 words

import pandas as pd
import re

# =====================================================
# CONFIGURATION
# =====================================================

INPUT_CSV = ".csv" # Stage 1 or Stage 2 csv
OUTPUT_CSV = "output_with_wordcount.csv"

# =====================================================
# WORD COUNT FUNCTION
# =====================================================

# URLs
URL_PATTERN = re.compile(
    r"(?:https?://[^\s]+|ftp://[^\s]+|www\.[^\s]+)",
    flags=re.IGNORECASE
)

# Emails
EMAIL_PATTERN = re.compile(
    r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b",
    flags=re.IGNORECASE
)

# Normal words/numbers with internal hyphens
WORD_PATTERN = re.compile(
    r"\b[A-Za-z0-9]+(?:-[A-Za-z0-9]+)*\b"
)

def count_words(text):
    if pd.isna(text):
        return 0

    text = str(text)

    # Count URLs
    urls = URL_PATTERN.findall(text)
    url_count = len(urls)
    text = URL_PATTERN.sub(" ", text)

    # Count emails
    emails = EMAIL_PATTERN.findall(text)
    email_count = len(emails)
    text = EMAIL_PATTERN.sub(" ", text)

    # Count remaining words
    normal_words = WORD_PATTERN.findall(text)

    return url_count + email_count + len(normal_words)

# =====================================================
# LOAD DATA
# =====================================================

df = pd.read_csv(INPUT_CSV)

# Check required columns
required_columns = ["text_analyzed"] # text_analyzed is the current "analysis" column but without the POS tags. Therefore, this code aims at counting the number of words from the final text excerpt analyzed previous to tagging (manual annotation)
missing = [col for col in required_columns if col not in df.columns]

if missing:
    raise ValueError(f"Missing required column(s): {', '.join(missing)}")

# =====================================================
# COMPUTE WORD COUNTS
# =====================================================

df["word_count2"] = df["text_analyzed"].apply(count_words) # word_count2 was only used to keep two word count variables, one for the word count of the entire post and another for the word count of analyzed data up to 500 words

# =====================================================
# SAVE RESULTS
# =====================================================

df.to_csv(OUTPUT_CSV, index=False)

# =====================================================
# PRINT RESULTS
# =====================================================

print(f"Processed {len(df)} rows.")
print(f"Output saved to: {OUTPUT_CSV}")

print("\nWord count by post:\n")

for index, row in df.iterrows():
    username = row["Redditor"]
    word_count = row["word_count2"]
    original_post = row["original_post"]

    print(f"Row {index + 1}")
    print(f"Username: {username}")
    print(f"Word count: {word_count}")
    print(f"Original post:\n{original_post}")
    print("-" * 60)

# =====================================================
# OPTIONAL: SHOW SAMPLE OF DATAFRAME
# =====================================================

print("\nFirst five rows:")
print(df[["text_analyzed", "word_count2"]].head())
```
