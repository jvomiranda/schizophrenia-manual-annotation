```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_validate
from boruta import BorutaPy

# Load data

df = pd.read_csv("stage1.csv")

# Labels
y = df["group"].values

# Linguistic features (12th column onwards)
X = df.iloc[:, 11:]
print(X.head())
feature_names = X.columns
X = X.values

print(X.shape, y.shape)
print(X[:5])


# Random Forest Classification Performance

rf = RandomForestClassifier(
    n_estimators=1000,
    random_state=42,
    class_weight="balanced",
    n_jobs=-1
)

cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

scores = cross_validate(
    rf,
    X,
    y,
    cv=cv,
    scoring=[
        "accuracy",
        "precision",
        "recall",
        "f1",
        "roc_auc"
    ]
)

print("\n==============================")
print("Random Forest Performance")
print("==============================")

for metric in scores:
    if metric.startswith("test_"):
        print(
            f"{metric[5:]:<10}: "
            f"{scores[metric].mean():.3f} ± {scores[metric].std():.3f}"
        )

# Boruta Feature Selection

rf_boruta = RandomForestClassifier(
    n_estimators=1000,
    random_state=42,
    class_weight="balanced",
    n_jobs=-1
)

boruta = BorutaPy(
    estimator=rf_boruta,
    n_estimators="auto",
    random_state=42,
    verbose=2
)

boruta.fit(X, y)


# Results

confirmed = feature_names[boruta.support_]

weak = feature_names[boruta.support_weak_]

rejected = feature_names[
    ~(boruta.support_ | boruta.support_weak_)
]


# Print Results

print("\n==============================")
print("Confirmed Features")
print("==============================")

for f in confirmed:
    print(f)

print("\n==============================")
print("Weak (Tentative) Features")
print("==============================")

for f in weak:
    print(f)

print("\n==============================")
print("Rejected Features")
print("==============================")

for f in rejected:
    print(f)


# Rankings


ranking = pd.DataFrame({
    "Feature": feature_names,
    "Ranking": boruta.ranking_,
    "Confirmed": boruta.support_,
    "Weak": boruta.support_weak_
}).sort_values("Ranking")

print("\n==============================")
print("Feature Ranking")
print("==============================")

print(ranking)


# 7. Save results

ranking.to_csv("Boruta_Feature_Ranking.csv", index=False)

pd.DataFrame({
    "Confirmed": pd.Series(confirmed),
    "Weak": pd.Series(weak),
    "Rejected": pd.Series(rejected)
}).to_csv("Boruta_Selected_Features.csv", index=False)
```
